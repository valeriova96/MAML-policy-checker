```python
class AttnClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(AttnClassifier, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        self.linear_1 = nn.Linear(self.input_size, self.hidden_size)
        self.linear_2 = nn.Linear(384, self.hidden_size)
        self.attn = nn.Linear(self.hidden_size, self.hidden_size)
        self.W_c = nn.Linear(self.hidden_size * 2, self.hidden_size)
        self.convnet = nn.Sequential( 
            
            nn.Conv1d(in_channels=560,
                      out_channels=50, kernel_size=3),

            nn.ELU(),

            nn.Conv1d(in_channels=50, out_channels=50,
                      kernel_size=3, stride=2),

            nn.ELU())
        
        self.fc1 = nn.Linear(62, 10)

        self.fc2 = nn.Linear(500, output_size) 

    def forward(self, input, policy):
        input = self.linear_1(input) # batch_size x 128 x 280
        policy = self.linear_2(policy) # batch_size x policy_sentences x 280
        
        # Compute the transpose of the policy on the last 2 dimensions
        # keeping batch size where it is
        policy_t = torch.transpose(policy, 1, 2)
        alignment_scores = torch.bmm(self.attn(input), policy_t) # (batch_size x 128 x 280) x (batch_size_p x 280 x policy_sentences)
        attn_weights = F.softmax(alignment_scores, dim=-1) # batch_size x 128 x policy_sentences
        c_t = torch.bmm(attn_weights, policy) # batch_size x 128 x 280
        output = torch.cat([input, c_t], dim=-1) # batch_size x 128 x 560

        output = output.permute(0, 2, 1)
        features = self.convnet(output)
        
        intermediate_vector = self.fc1(features)
        intermediate_vector = intermediate_vector.reshape(1, -1)
        prediction_vector = self.fc2(intermediate_vector)

        return prediction_vector
```

```python
def get_original_label(label, classes, train=True):
    
    if(train):
        classes = classes[:4]
        l = classes[label]
        return l
    else:
        classes = classes[4:]
        l = classes[label]
        return l
```

### Construct the model and define loss and accuracy functions as well as hyperparameters

```python
def accuracy(predictions, targets):
    predictions = torch.tensor([torch.argmax(p) for p in predictions])
    acc = (predictions == targets).sum().float()
    acc /= len(targets)
    return acc.item()
```

```python
lr=0.0001
maml_lr=0.0001


model = AttnClassifier(768, 280, ways)
model = model.to(device)
meta_model = l2l.algorithms.MAML(model, allow_nograd=True, lr=maml_lr)
opt = optim.Adam(meta_model.parameters(), lr=lr)
loss_func = nn.CrossEntropyLoss()
accuracy = accuracy
```

```python
class _BatchedDataset(torch.utils.data.Dataset):
    def __init__(self, batched):
        self.sents = [s for s in batched[0]]
        self.ys = [y for y in batched[1]]
    
    def __len__(self):
        return len(self.ys)
    
    def __getitem__(self, idx):
        return (self.sents[idx], self.ys[idx])
```

### Define useful functions and training loop

*collate_tokens* converts our batch of sentences (a list essentially), into a 2d tensor *batch_size* x *length_of_sentence*, w/ *length_of_sentence* equal for all the elements of the batch (we use padding).

```python
def collate_tokens(values):
    size = 128

    values = pad_sequences(
    values, 
    maxlen=size, 
    dtype="long", 
    truncating="post", 
    padding="post",
    value=0)

    return torch.tensor(values)
```

#### Experiment

```python
def compute_loss(task, policies_dict, learner, loss_func, accuracy_func, batch=1):
    loss = 0.0
    acc = 0.0
    steps = 0
    for i, (tweet, label) in enumerate(torch.utils.data.DataLoader(
            _BatchedDataset(task), batch_size=batch, shuffle=True, num_workers=0)):
        
        # Convert sentences into RoBERTa embedding vectors
        tweet = collate_tokens([roberta.encode(sent) for sent in tweet])
        with torch.no_grad():
            tweet = roberta.extract_features(tweet)

        policy = policies_dict[label.item()]

        # Move to GPU
        policy = policy.unsqueeze(0).to(device)
        tweet, label = tweet.to(device), label.view(-1).to(device)
    
        output = learner(tweet, policy)
        output = output.to(device)

        curr_loss = loss_func(output, label)
        loss += curr_loss
        label = label.to('cpu')
        curr_acc = accuracy_func(output, label)
        acc += curr_acc
        steps += 1
    loss /= steps
    acc /= steps
    return loss, acc
```

```python
from tqdm import tqdm

iterations=60
tps=8
fas=1

tqdm_bar = tqdm(range(iterations))

val_accs = []
tr_losses = []

tr_policies_ready = False
val_policies_ready = False
tr_policies = dict()
val_policies = dict()

for _ in tqdm_bar:
    iteration_error = 0.0
    iteration_acc = 0.0
    train_error = 0.0

    for _ in range(tps):
        learner = meta_model.clone()
        train_task, valid_task = train_gen.sample(), validation_gen.sample()


        # Fast Adaptation
        for _ in range(fas):

            if (not tr_policies_ready):
                _, labels = train_task
                labels = set([l.item() for l in labels])

                for l in labels:
                    label = l
                    # retrieve the policiy
                    original_l = get_original_label(label, classes)
                    curr_policy = policies[original_l]
                    # store the policy in the dictionary
                    tr_policies[label] = curr_policy
                tr_policies_ready = True


            train_error, _ = compute_loss(train_task, tr_policies, learner, loss_func, accuracy)
            tr_losses.append(train_error)
            learner.adapt(train_error)
        

        
        # encode the policies for validation
        if (not val_policies_ready): 
            _, labels = valid_task
            labels = set([l.item() for l in labels])
            for l in labels:
                label = l
                # retrieve the policiy
                original_l = get_original_label(label, classes, train=False)
                curr_policy = policies[original_l]
                # store the policy in the dictionary
                val_policies[label] = curr_policy
            val_policies_ready = True
        
        # Compute validation loss
        valid_error, valid_acc = compute_loss(valid_task, val_policies, learner, loss_func, accuracy)
        iteration_error += valid_error
        iteration_acc += valid_acc

    iteration_error /= tps
    iteration_acc /= tps
    tqdm_bar.set_description(f'Loss : {iteration_error.item()} Acc : {iteration_acc}')
    val_accs.append(iteration_acc)
    # Take the meta-learning step
    opt.zero_grad()
    iteration_error.backward()
    opt.step()
    
print (f'first and best validation accuracy: ->{val_accs[0]} ->{max(val_accs)}')
```
